```python
def table_info(conn,cursor):
    """
    prints out all of the columns of every table in the DB
    
    conn: database connectin object
    cursor: cursor object
    """
    tables= cur.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
    for table_name in tables:
        table_name=table_name[0]
        table=pd.read_sql_query("SELECT* from {} LIMIT 0".format(table_name), conn)
        print(table_name)
        for col in table.columns:
            print("\t"+ col)
            print()
```

```python
# Create and train the Random Forest Regressor model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)
```

```python
# Calculate the evaluation metrics for Random Forest Regressor
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
print("Random Forest Regressor Mean Squared Error:", mse_rf)
print("Random Forest Regressor R-squared:", r2_rf)
```
